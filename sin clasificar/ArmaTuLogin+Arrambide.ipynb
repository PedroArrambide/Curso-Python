```python
BD = {}

def registrar_usuario():
    usuario = input("Ingrese el nombre de usuario: ")
    contraseña = input("Ingrese la contraseña: ")
    BD[usuario] = contraseña
    print("¡Usuario registrado con éxito!")

def iniciar_sesion():
    usuario = input("Ingrese su nombre de usuario: ")
    contraseña = input("Ingrese su contraseña: ")
    if usuario in BD and BD[usuario] == contraseña:
        print("¡Inicio de sesión exitoso!")
    else:
        print("Nombre de usuario o contraseña incorrectos.")

def mostrar_datos():
    print("Datos almacenados en el diccionario:")
    for usuario, contraseña in BD.items():
        print(f"Usuario: {usuario} - Contraseña: {contraseña}")

while True:
    print("1. Registrar usuario")
    print("2. Iniciar sesión")
    print("3. Ver datos registrados")
    print("4. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        registrar_usuario()
    elif opcion == "2":
        iniciar_sesion()
    elif opcion == "3":
        mostrar_datos()
    elif opcion == "4":
        break
    else:
        print("Opción inválida. Por favor, seleccione una opción válida.")
```
